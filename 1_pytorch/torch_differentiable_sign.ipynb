```python
import torch
```

```python
def differentiable_sign(x, epsilon=1e-6):
    return torch.tanh(x / epsilon)

n = 50
for i in range(2*n) : 
    x = torch.tensor(-1 + i/n)*10
    print("{} : {:.3f} -> {}".format(i,x,differentiable_sign(x)))
```

```python
def differentiable_abs(x, epsilon=1e-6):
    return torch.sqrt(x ** 2 + epsilon ** 2)

n = 50
for i in range(2*n) : 
    x = torch.tensor(-1 + i/n)*10
    print("{} : {:.3f} -> {}".format(i,x,differentiable_abs(x)))
```
