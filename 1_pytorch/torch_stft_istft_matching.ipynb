```python
x = torch.rand(1,64000)
X = torch.stft(x,n_fft=512,center=True)
print(X.shape)
Z = X[:,:,:500,:]
print(Z.shape)
y = torch.istft(X,n_fft=512)
z = torch.istft(Z,n_fft=512)
print(y.shape)
# reducing target length due to STFT 1 frame mismatch
print(torch.sum(x[:,:-128]-z))
print(torch.sum(x-y))
```

```python
x = torch.randn(1,16000)
y = torch.stft(x,n_fft=512,return_complex=True,pad_mode="circular")
z = torch.istft(y,n_fft=512)

print(x.shape)
print(z.shape)

torch.sum(torch.abs(x-z))

import matplotlib.pyplot as plt    

plt.figure()
plt.plot(x[0])
plt.title("x")

plt.figure()
plt.plot(z[0])
plt.title("z")

plt.figure()
plt.plot(torch.abs(x-z)[0])
plt.title("x-z")
```
