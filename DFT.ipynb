```python
import numpy as np
import librosa as rs

import matplotlib.pyplot as plt   
import IPython.display as ipd
```

```python
def DFT(signal, N = 512):
    N = len(signal)
    spectrum = np.zeros(N, dtype=np.complex128)
    for k in range(N):
        spectrum[k] = sum(signal[n] * np.exp(-2j * np.pi * k * n / N) for n in range(N))
    return spectrum

def iDFT(spectrum, N = 512):
    N = len(spectrum)
    signal = np.zeros(N, dtype=np.complex128)
    for n in range(N):
        signal[n] = sum(spectrum[k] * np.exp(2j * np.pi * k * n / N) for k in range(N)) / N
    return signal
```

```python
path = "./data/female_1.wav"
```

```python
x = rs.load(path,sr=16000)[0]
print(x.shape)
plt.figure()
plt.specgram(x,NFFT=512, Fs=16000)

X = []
idx = 0
while idx < 1000:
    tmp = DFT(x[idx:idx+512])
    X.append(tmp)
    idx += 128
X = np.array(X)
print(X.shape)

```
