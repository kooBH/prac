```python
import librosa as rs
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt   
import IPython.display as ipd
```

```python
sr = 16000
clean, _ = rs.load("data/female_1.wav",sr=sr)
rir,_ = rs.load("data/RIR_SLR26_med_001_00001.wav",sr=sr)
reverbed = signal.fftconvolve(clean, rir)


plt.figure()
plt.plot(clean)
plt.title("clean")
ipd.display(ipd.Markdown('## clean'))
ipd.display(ipd.Audio(clean, rate=sr))

plt.figure()
plt.plot(rir)
plt.title("RIR")

plt.figure()
plt.plot(reverbed)
plt.title("Reverbed")
ipd.display(ipd.Markdown('## Reverbed'))
ipd.display(ipd.Audio(reverbed, rate=sr))
```

```python
rir_peak = np.argmax(rir)
print(rir_peak)

reverbed_peak = signal.fftconvolve(clean, rir[:rir_peak])

plt.figure()
plt.plot(reverbed_peak)
plt.title("reverbed_peak")
ipd.display(ipd.Markdown('## reverbed_peak'))
ipd.display(ipd.Audio(reverbed_peak, rate=sr))
```

```python
reverbed_1500 = signal.fftconvolve(clean, rir[:1000])

plt.figure()
plt.plot(reverbed_peak)
plt.title("reverbed_1500")
ipd.display(ipd.Markdown('## reverbed_1500'))
ipd.display(ipd.Audio(reverbed_1500, rate=sr))
```
