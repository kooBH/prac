```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
cpp_feat = np.fromfile("out.bin", dtype=np.float32)
py_feat = np.fromfile("x.npy", dtype=np.float32)

print(cpp_feat.shape)
print(py_feat.shape)

error_level_rtol = 1e-03
error_level_atol = 1e-03

print("TESTING torch vs ONNX")
try :
    np.testing.assert_allclose(cpp_feat, py_feat, rtol=error_level_rtol, atol=error_level_atol,verbose=True)
except AssertionError :
    print(f"ERROR is larger than rtol {error_level_rtol} | atol {error_level_atol}")

print(cpp_feat[:5])
print(py_feat[:5])


plt.figure()
plt.plot(cpp_feat)
plt.plot(py_feat)
plt.plot(np.abs(py_feat-cpp_feat),linestyle='dashed')
plt.legend(["cpp","py","diff"])
```
