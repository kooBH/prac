```python
import torch
import matplotlib.pyplot as plt   
import IPython.display as ipd
```

```python
window_length = 400
win = torch.hamming_window(window_length=window_length)

alpha = 0.54
beta =0.46
win_impl = torch.zeros(window_length)
for i in range(window_length) : 
    win_impl[i] = alpha - beta * torch.cos(torch.tensor(2.0 * torch.pi * (i / window_length)));
print(win)

plt.figure()
plt.plot(win)
plt.plot(win_impl)
plt.legend(["torch","impl"])

plt.figure()
plt.plot(torch.abs(win-win_impl))
```
