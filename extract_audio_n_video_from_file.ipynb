```python
video = VideoFileClip('/home/data2/kbh/LRS2/main/5535415699068794046/00001.mp4')
print(video.fps)
audio = video.audio
print(audio.fps)
vid = []
for frame in video.iter_frames() :
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    vid.append(gray)
vid = np.array(vid)
print(vid.shape)

arr = audio.to_soundarray()
print(arr.shape)

ipd.display(ipd.Audio(arr[:,], rate=audio.fps))
```

```python
vid = []
cap = cv2.VideoCapture('/home/data2/kbh/LRS2/main/5535415699068794046/00001.mp4')
while(cap.isOpened()):
    ret, frame = cap.read()
    if not ret :
        break
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    vid.append(gray)
cap.release()

vid = np.array(vid)
print(vid.shape)
```
