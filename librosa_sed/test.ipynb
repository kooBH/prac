```python
import librosa 
import numpy as np
import matplotlib.pyplot as plt 
import os,glob
```

```python
x,_ = librosa.load("ding.wav",mono=True,sr=16000)

plt.figure()
plt.plot(x)
plt.title("x")
```

```python
for i in [5,8,10,15,20] : 
    temp,idx = librosa.effects.trim(x,top_db=i)
    plt.figure()
    plt.plot(temp)
    plt.title("{} {} {}".format(i,len(temp),idx))
```
