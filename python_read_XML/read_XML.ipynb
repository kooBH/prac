```python
import xml.etree.ElementTree as ET

# XML 파일을 파싱하고 루트 요소 가져오기
tree = ET.parse('meetings.xml')
root = tree.getroot()

# 루트 요소 이름 출력
print(root.tag)
print("----------------------------------")

data={}
# 루트 요소의 모든 자식 요소 출력
for meeting in root:
    #print(meeting.tag, meeting.attrib)
    obs = meeting.attrib.get('observation')
    #print(obs)
    info = {}
    for idx, speaker in enumerate(meeting.findall('speaker')) : 
        #print(speaker.tag, speaker.attrib)
        info[speaker.attrib.get('channel')] = speaker.attrib.get('global_name')
    #print(info)
    info = dict(sorted(info.items()))
    data[obs] = info
    
sorted_dict = dict(sorted(data.items()))

print(sorted_dict)
```
