librosa.effects.time_stretch(y: ndarray, *, rate: float, **kwargs: Any)→ ndarray[source]¶
Time-stretch an audio series by a fixed rate.

Parameters
ynp.ndarray [shape=(…, n)]
audio time series. Multi-channel is supported.

ratefloat > 0 [scalar]
Stretch factor. If rate > 1, then the signal is sped up. If rate < 1, then the signal is slowed down.

**kwargsadditional keyword arguments.
See librosa.decompose.stft for details.

Returns
y_stretchnp.ndarray [shape=(…, round(n/rate))]

````python
import librosa as rs
import numpy as np
import IPython.display as ipd
import pyrubberband as pyrb
````

# Librosa simple stretch

````python
sr = 16000

original = rs.load("data/male_1.wav",sr=sr)[0]
print(original.shape)

fast = rs.effects.time_stretch(original, rate=0.92)
print(fast.shape)

ipd.display(ipd.Markdown('## oriinal'))
ipd.display(ipd.Audio(original, rate=sr))

ipd.display(ipd.Markdown('## fast'))
ipd.display(ipd.Audio(fast, rate=sr))
````

# pyrubberband

````python
sr = 16000

original = rs.load("data/male_1.wav",sr=sr)[0]
print(original.shape)

fast = pyrb.time_stretch(original, sr, 0.92)
print(fast.shape)

ipd.display(ipd.Markdown('## oriinal'))
ipd.display(ipd.Audio(original, rate=sr))

ipd.display(ipd.Markdown('## fast'))
ipd.display(ipd.Audio(fast, rate=sr))
````
