```python
import glob,os
import librosa
import numpy as np

import matplotlib.pyplot as plt   
import IPython.display as ipd
```

```python
x,sr = librosa.load("sample.m4a",sr=16000)
```

```python
print(x.shape)
print(sr)
```

```python
ipd.display(ipd.Markdown('## input'))
ipd.display(ipd.Audio(x, rate=sr))

plt.figure()
plt.plot(x)
plt.title("x")
```

https://librosa.org/doc/main/generated/librosa.effects.trim.html

```python
y = librosa.effects.split(x,top_db=40)
print(len(y))
print(y)
```

```python
cnt = 0 

bool_audio = False
bool_plot = False
for i in range(len(y)) :
    if y[i,1] - y[i,0] < 4000:
        continue
    if np.mean(np.abs(x[y[i,0]:y[i,1]])) < 0.05:
        continue
    if bool_plot :
        plt.figure()
        plt.plot(x[y[i,0]:y[i,1]])
        plt.title(i)
    
    if bool_audio : 
        ipd.display(ipd.Markdown(str(i)))
        ipd.display(ipd.Audio(x[y[i,0]:y[i,1]], rate=sr))
    #print(y[i,1]) 
    #print(y[i,0])
    
    cnt += 1
    
print(cnt)
```
